# Work on the Numerical attributes -

```python
train_df_numerical = train_df[numerical_attributes]

# Test operations
test_df_numerical = test_df[numerical_attributes]
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')

imputer = imputer.fit(train_df_numerical)
train_df_numerical_filled = pd.DataFrame(imputer.transform(train_df_numerical))
train_df_numerical_filled.columns = train_df_numerical.columns
train_df_numerical_filled.index = train_df_numerical.index


# Test operations
imputer = imputer.fit(test_df_numerical)
test_df_numerical_filled = pd.DataFrame(imputer.transform(test_df_numerical))
test_df_numerical_filled.columns = test_df_numerical.columns
test_df_numerical_filled.index = test_df_numerical.index
```

```python
# train_df_numerical_filled
test_df_numerical_filled
```

```python
# Scale the numerical data for better predictions
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Drop the SalePrice before scaling
# train_df_numerical_filled = train_df_numerical_filled.drop("SalePrice", axis=1)

scaler = scaler.fit(train_df_numerical_filled)
train_df_numerical_filled_scaled = pd.DataFrame(scaler.transform(train_df_numerical_filled))
train_df_numerical_filled_scaled.columns = train_df_numerical_filled.columns
train_df_numerical_filled_scaled.index = train_df_numerical_filled.index
train_df_numerical_filled_scaled

# Test operations
scaler = scaler.fit(test_df_numerical_filled)
test_df_numerical_filled_scaled = pd.DataFrame(scaler.transform(test_df_numerical_filled))
test_df_numerical_filled_scaled.columns = test_df_numerical_filled.columns
test_df_numerical_filled_scaled.index = test_df_numerical_filled.index
test_df_numerical_filled_scaled
```

# Work on the categorical attributes

```python
# One hot encode the categories
```

```python
train_df_categorical = train_df[categorical_attributes]
train_df_categorical.reset_index(drop=True, inplace=True)
train_df_categorical = train_df_categorical.fillna("empty")
len(set(train_df_categorical.columns))

# Test operations
test_df_categorical = test_df[categorical_attributes]
test_df_categorical.reset_index(drop=True, inplace=True)
test_df_categorical = test_df_categorical.fillna("empty")
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
train_df_categorical_encoded = pd.get_dummies(train_df_categorical, columns=list(train_df_categorical.columns))

# Test operations
test_df_categorical_encoded = pd.get_dummies(test_df_categorical, columns=list(test_df_categorical.columns))
```

```python
# Combine the cleaned dataframes and train models and predict
train_df_combined = pd.concat([train_df_numerical_filled_scaled,train_df_categorical_encoded], axis=1)
train_df_combined

test_df_combined = pd.concat([test_df_numerical_filled_scaled,test_df_categorical_encoded], axis=1)
test_df_combined

for category in list(set(train_df_combined) - set(test_df_combined)):
    test_df_combined[category] = 0

for category in list(set(test_df_combined) - set(train_df_combined)):
    train_df_combined[category] = 0

# list(set(test_df_combined.columns)- set(train_df_combined.columns))
```

# Train the models

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

train_set_X, test_set_X, train_set_Y, test_set_Y = train_test_split(train_df_combined, train_df_y, test_size=0.2, random_state=42)
```

```python
lr_model = LinearRegression()
lr_model.fit(train_df_combined, train_df_y)
```

```python
predictions = lr_model.predict(train_df_combined)
```

```python
# Calculate the mean squared error
import numpy as np
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(train_df_y, predictions)
# Print RSME - Root(mean squared error)
print(np.sqrt(mse))
```
