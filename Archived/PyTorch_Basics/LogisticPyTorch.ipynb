```python
# Pytorch basic code for running logistic regression
```

```python
import pandas as pd
df = pd.read_csv("logistic_regression_ready_data_insurance_charges.csv")
df
```

```python
import torch
import torch.nn as nn
```

```python
from torch.utils.data import DataLoader, TensorDataset
train_dataset = TensorDataset(torch.tensor(df.drop("charges", axis=1).values), torch.tensor(df.charges.values))
train_dataloader = DataLoader(train_dataset, batch_size=4, shuffle=True)
train_dataloader
next(iter(train_dataloader))
```

```python
import torch.nn.functional as F
loss_function = F.mse_loss
```

```python
linear_regression_model = nn.modules.Linear(in_features=df.shape[1] - 1, out_features=1)
```

```python
# Define optimizer
opt = torch.optim.Adam(linear_regression_model.parameters(), lr=1e-5)
```

```python
for epoch in range(100):
    for x_batch, y_batch in train_dataloader:
        predictions = linear_regression_model(x_batch.float())
        loss = loss_function(predictions, y_batch.float())
        loss.backward()
        opt.step()
        opt.zero_grad()
    print("Loss : ", loss_function(linear_regression_model(torch.tensor(df.drop("charges", axis=1).values).float()), torch.tensor(df.charges.values).float()))
```
