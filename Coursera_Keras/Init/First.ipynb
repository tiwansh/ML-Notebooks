```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

# Prepare the df

```python
import pandas as pd
```

```python
df = pd.read_csv("insurance.csv")
```

```python
df
```

```python
df.sex = df.sex.replace({"female" : 0, "male" :1})
df.smoker = df.smoker.replace({"yes" : 0, "no" : 1})
df_y = df.charges
df_sex = df.sex
df_smoker = df.smoker
df.drop(['region', 'charges', 'sex', 'smoker'], inplace=True, axis=1)
df
```

```python
# Scale all the features, except smoker
```

```python
from sklearn.preprocessing import StandardScaler
df = StandardScaler().fit_transform(df)
df
```

```python
df_x = pd.DataFrame(data=df, columns=['age', 'bmi', 'children'])
df_x['sex'] = df_sex
df_x['smoker'] = df_smoker
df_x
```

```python
df_y
```

```python
df_x
```

# Train the keras model

```python
model.add(Dense(100, activation="relu", input_shape=(df_x.shape[1], )))
model.add(Dense(10, activation="relu"))
# model.add(Dense(100, activation="relu"))
model.add(Dense(1))
```

```python
model.compile(optimizer="adam", loss="mean_squared_error")
model.fit(df_x, df_y, validation_split=0.2, epochs=100, verbose=2)
```

```python
predictions = model.predict(df_x)
predictions
```

```python
df_y
```

```python
from keras.metrics import RootMeanSquaredError
```

```python
m = RootMeanSquaredError()
m.update_state(predictions, df_y)
```

```python
m.result().numpy()
```
