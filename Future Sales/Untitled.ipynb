```python
# Average sale of an item in each shop for a month
```

```python
df_train = df_sales_train.groupby(["shop_id", "date_block_num", "item_id"])["item_cnt_day"].sum().reset_index()
```

```python
# df_train
```

```python
price_map = dict()
for index, row in df_sales_train.iterrows():
    price_map[row['item_id']] = row['item_price']
price_map
```

```python
df_train
```

```python
df_train['item_price'] = df_train['item_id'].apply(lambda x : price_map.get(x))
```

```python
df_train
```

```python
df_train[(df_train.shop_id == 0) & (df_train.item_id == 35)]
```

```python
df_train_final = df_train.groupby(['shop_id', 'item_id'], as_index=False)['item_cnt_day'].mean()
df_train_final
```

```python
# pd.get_dummies(df_train_final)
```

```python
df_train_final['item_id'] = df_train_final['item_id'].astype(str)
df_train_final['shop_id'] = df_train_final['shop_id'].astype(str)
```

```python
df_train_final['item_id'] = "item" + df_train_final['item_id'].astype(str)
df_train_final['shop_id'] = "shop" + df_train_final['shop_id'].astype(str)
```

```python
df_train_final
```

```python
df_final = pd.get_dummies(df_train_final)
```
